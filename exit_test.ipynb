**split train and test data**

```python
X=train_data.drop('selection',axis=1)
y=train_data['selection']
```

```python
Xtest = test_data
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state=42, test_size=0.2)
```

**logistic regression**

```python
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.linear_model import LogisticRegression
log  = LogisticRegression()
from sklearn.model_selection import train_test_split
```

```python
log.fit(X_train,y_train)
y_pred_lr = log.predict(X_test)
y_pred_lr
```

```python
#import confusion matrix,precision score,accuracy score,recall score
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score, recall_score
```

```python
#confusion matrix
confusion_matrix(y_test,y_pred_lr)
```

```python
#precision score
precision_score(y_test,y_pred_lr,pos_label=1,average='micro')
```

```python
#recallscore
recall_score(y_test,y_pred_lr,pos_label=1,average='micro')
```

```python
#check accuracy
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred_lr)
```

**kNN**

```python
#import KNeighborsClassifier
from sklearn.neighbors import KNeighborsClassifier
```

```python
metric_k=[]
neighbors=range(3,15)
for k in neighbors:
  classifier= KNeighborsClassifier(n_neighbors=k)
  classifier=classifier.fit(X_train,y_train)
  y_pred_kNN=classifier.predict(X_test)
  acc=accuracy_score(y_test,y_pred_kNN)
  metric_k.append(acc)
```

```python
plt.plot(neighbors,metric_k,'o-')
plt.xlabel('k values')
plt.ylabel('Accuracies')
plt.grid()
```

```python
classifier= KNeighborsClassifier(n_neighbors=5)
classifier=classifier.fit(X_train,y_train)
y_pred_kNN=classifier.predict(X_test)
```

```python
#find confusion matrix
confusion_matrix(y_test,y_pred_kNN)
```

```python
#find precision score
precision_score(y_test,y_pred_kNN,pos_label=1,average='micro')
```

```python
#find recall score
recall_score(y_test,y_pred_kNN,pos_label=1,average='micro')
```

```python
#find accuracy score
accuracy_score(y_test,y_pred_kNN)
```
